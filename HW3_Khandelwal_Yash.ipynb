````
# Type solution here
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# read the csv file
avocado_File = pd.read_csv('avocado.csv') 

# store into a dataframe and show columns
avocado = pd.DataFrame(avocado_File, columns=["Date", "AveragePrice", "Total Volume"])

# change column to timestamp
avocado["Date"] = pd.to_datetime(avocado["Date"])

# sort the date column
avocado.sort_values(by = "Date", inplace=True)


# multiply averageprice and total volume to get total revenue
totalRevenue = avocado["AveragePrice"]*avocado["Total Volume"]

# create a new column for totalrevenue 
avocado.insert(3, "TotalRevenue", totalRevenue, True)

# group together the dataframe over the date and create a new dataframe called avocado1
avocado1=avocado.groupby('Date').sum()

# find the average price by dividing total revenue by total volume
avocado1['AveragePrice'] = avocado1['TotalRevenue']/avocado1['Total Volume']


# Rolling with a window of 12 observations. We then consider mean of those 12 observations
avocado1["AveragePriceMean"] = avocado1['AveragePrice'].rolling(20).mean()
avocado1["Total Volume Mean"] = avocado1["Total Volume"].rolling(20).mean()

# grab the data and store it in variables
x = avocado1.index.get_level_values('Date')
y = avocado1["AveragePriceMean"]
y2 = avocado1["Total Volume Mean"]


myFig = plt.figure(figsize=(20,10))
myFig.suptitle('Avocado Prices and Volume Time Series', size = 16)

ax1 = myFig.add_subplot(1,2,1)
ax2 = myFig.add_subplot(1,2,2)

# plot for average prices
ax1.plot(x, y, color = "b", marker = 'o')
ax1.set_ylabel("Average Price")
ax1.tick_params(axis='x', labelrotation = 90)

# plot for total volume
ax2.plot(x,y2,color = "b", marker = 'o')
ax2.set_ylabel("Total Volume")
ax2.tick_params(axis='x', labelrotation = 90)
````
